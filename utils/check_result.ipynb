```python
import pandas as pd

csv1_paph = "/workspace/outputs/2024-10-01/16-19-30/result.csv"
csv2_paph = "/workspace/outputs/2024-10-01/16-32-36/result.csv"
csv3_paph = "/workspace/outputs/2024-10-01/16-40-06/result.csv"

df1 = pd.read_csv(csv1_paph)
df2 = pd.read_csv(csv2_paph)
df3 = pd.read_csv(csv3_paph)
```

```python
df1 = df1.rename(columns={"category": "category-gpt4o", "reason": "reason-gpt4o"})
df2 = df2.rename(columns={"category": "category-gemini-flash", "reason": "reason-gemini-flash"})
df3 = df3.rename(columns={"category": "category-gemini-pro", "reason": "reason-gemini-pro"})


df = pd.concat([df1, df2, df3], axis=1)
df = df.loc[:,~df.columns.duplicated()]
```
